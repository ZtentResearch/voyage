```python
import os
```

```python
cat_dir = "./data/image-cats/"
png_files = [f for f in os.listdir(cat_dir) if f.endswith('.png')]
print(png_files)
batch = torch.zeros(len(png_files), 3, 256, 256, dtype=torch.uint8)
batch.shape
```

```python
for i, filename in enumerate(png_files):
    temp_image_array = imageio.imread(os.path.join(cat_dir, filename))
    image_tensor = torch.from_numpy(temp_image_array)
    image_tensor = image_tensor.permute(2, 0, 1)
    print(image_tensor.shape)
    image_tensor = image_tensor[:3]
    batch[i] = image_tensor
```

```python
batch = batch.float()
batch /= 255.0
```
