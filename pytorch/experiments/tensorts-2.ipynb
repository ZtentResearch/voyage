```python
import torch
```

```python
points = torch.tensor([[4.0, 1.0], [5.0, 3.0], [2.0, 1.0]])
points
```

```python
points_t = points.t()
points_t
```

```python
id(points_t.storage()) == id(points.storage())
```

```python
points.stride(), points_t.stride()
```

```python
# So transposing is just changing the order of stride
```

```python
a = torch.tensor([[[1,2,3,4], [1,2,3,6]],
                  [[0,2,1,4], [7,2,3,6]]])
a.transpose(0, 1)
```

```python
points.is_contiguous(), points_t.is_contiguous()
```

```python
points_t_cont = points_t.contiguous()
points_t_cont.storage()
```
