```python
# _like -> Basically 
# Every _like function does one thing: 
# copy the shape, dtype, and device of an existing tensor, then fill the new tensor using whatever rule its base function uses.
torch.zeros_like(x)   # same shape/dtype/device as x, filled with 0
torch.ones_like(x)    # same shape/dtype/device as x, filled with 1
torch.rand_like(x)    # same shape/dtype/device as x, filled uniform [0,1)
torch.randn_like(x)   # same shape/dtype/device as x, filled standard normal
torch.empty_like(x)   # same shape/dtype/device as x, uninitialized garbage
torch.full_like(x, 5) # same shape/dtype/device as x, filled with value v
```

```python
x = torch.tensor([[1.0, 2.0], [3.0, 4.0]], dtype=torch.float64, device='cuda')

# The manual way is verbose and easy to forget attributes
# zeros_manual = torch.zeros(size=x.shape, dtype=x.dtype, device=x.device)

# The _like way (clean and readable)
zeros_like = torch.zeros_like(x)
zeros_like.shape, zeros_like.dtype, zeros_like.device
```

```python
# Boolean mask

scores = torch.tensor([0.2, 0.85, 0.4, 0.95, 0.1])
mask = scores > 0.5 
print(mask)

# Apply the mask to extract values
confident_scores = scores[mask]

print(confident_scores)
```

```python
xp = torch.tensor([1.0, -2.0, 3.0, -4.0])

relu_like = torch.where(xp > 0, xp, 0.0) # if xp > 0 is true then it picks xp otherwise 0.0, and it does not change the tensor shape

print(relu_like)
```

```python
# Using tensor index /fancy indexing 
embeddings = torch.tensor([
    [0.1, 0.1, 0.1],  
    [0.5, 0.5, 0.5],  
    [0.9, 0.9, 0.9],  
    [0.3, 0.3, 0.3],  
])

# Retrieve embeddings for word batch: [Word 2, Word 0, Word 2, Word 1]
indices = torch.tensor([2, 0, 2, 1])
batch_embeddings = embeddings[indices]

print(batch_embeddings)
```

```python
# Gather
x = torch.tensor([
    [10, 20, 30],
    [40, 50, 60]
])

index = torch.tensor([
    [2],
    [0]
])

out = torch.gather(x, dim=1, index=index)

print(out)
```

```python
# scatter
out = torch.zeros(2, 3, dtype=torch.int64)

"""
[[0,0,0],
[0,0,0]]
"""

src = torch.tensor([
    [5, 6],
    [7, 8]
])

index = torch.tensor([
    [1, 2],
    [0, 2]
])

out.scatter_(1, index, src)

print(out)
```
