### How to install torch with uv?

- Start an uv project

- Then on the `pyproject.toml` file add these config for proper wheel

```
[[tool.uv.index]]
name = "pytorch-cu128"
url = "https://download.pytorch.org/whl/cu128"
explicit = true

[tool.uv.sources]
torch = [{ index = "pytorch-cu128" }]
torchvision = [{ index = "pytorch-cu128" }]
```
Here I pointed the wheel location on the `url` and named it `pytorch-cu128` (<=cuda version on `nvidia-smi`).

Then I linked torch and torchvision package with it as uv uses this url for these 2 package

- Then just do `uv add torch torchvision`. We will get our own cuda runtime all we need in our device is the nvidia driver

- Torch is a `Eager Mode` library because when we write code or implement model we can run it just like python

  We don't need to compile them and then run it is just like python and under the hood the backends(c++,cuda,etc) takes care of everything.

- SO over all
<b>`torch.tensor` carries the data -> `torch.nn` for defining the models -> autograd compute the gradients -> `torch.optim` update the weights</b> 

- `torch.compile` is for actual finalized training run which leave the eager mode and does JIT compiled run

- `torch.distributed` is for multi-GPU run
