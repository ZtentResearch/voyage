```python
x = torch.tensor([[1], [2], [3]])         # shape [3, 1]

# expand – broadcasting style (almost free)
y = x.expand(3, 4)  # y is a view. Memory still only stores the original 3 values.
print(y, y.shape)

# repeat – actually copies data
z = x.repeat(1, 4) # z owns its own storage with the values repeated.
print(z, z.shape)
```

```python
x = torch.tensor([[1., 2., 3.],
                  [4., 5., 6.]]) 
sm = x.sum()
print(sm, sm.shape) # tensor(21.)

s_dim0 = x.sum(dim=0) 
print(s_dim0)        
print(s_dim0.shape)  

# keepdim=True: Leaves the dimension as size 1
s_keep = x.sum(dim=0, keepdim=True)
print(s_keep)        
print(s_keep.shape)  
```

```python
v = torch.tensor([3.0, 4.0])

# sqrt(3^2 + 4^2) = sqrt(9 + 16) = sqrt(25) = 5
print(torch.linalg.norm(v)) # tensor(5.)

# we can also do this along dimensions for batches of vectors
batch = torch.tensor([[3., 4.], [6., 8.]])
print(torch.linalg.norm(batch, dim=1)) 
```

```python
# Matmul
A = torch.tensor([[1, 2], 
                  [3, 4]])

B = torch.tensor([[5, 6],
                  [7, 8]])

# Element-wise (top-left is 1*5)
print(A * B) 
# [[ 5, 12],
#  [21, 32]]

# Matrix Multiplication (top-left is (1*5) + (2*7) = 19)
print(A @ B)
# [[19, 22],
#  [43, 50]]
```

#### Einsum rules

## The 3 Rules of Einsum

Every `einsum` string looks like this: `"inputs -> output"`.
(e.g., `"ij, jk -> ik"`)

1. **Letters are dimensions:** Each letter represents one axis of your tensor. A 2D matrix gets two letters (e.g., `ij`). A 3D tensor gets three (e.g., `bjk`).
2. **Repeating a letter across inputs means MULTIPLY:** If input A has a `j` and input B has a `j`, PyTorch will multiply the elements along that `j` dimension.
3. **Omitting a letter in the output means SUM (Squash):** If a letter appears in the inputs but is missing on the right side of the `->`, PyTorch will sum up all the values along that dimension.


```python
# einsum() is crazy
"""
Matrix Multiplication (The Classic)
What we have: A is ij, B is jk (Matrix A's columns j match Matrix B's rows j)
What we want: ik (Matrix A's rows, Matrix B's columns)
How it works: j is repeated, so we multiply along it. j is missing from the output, 
so we sum those products. (This is exactly how you calculate a dot product for every row/col pair in standard matrix math).
"""
A = torch.tensor([[1, 2], [3, 4]])
B = torch.tensor([[5, 6], [7, 8]])

print(torch.einsum('ij, jk -> ik', A, B))
```

```python
# ---------------------------------------------------------
# SETUP: Create one 3D tensor for both examples
# Shape: [Batch=2, Sequence=4, Features=6]
# ---------------------------------------------------------
x_3d = torch.arange(48).reshape(2, 4, 6)
print(f"Original tensor shape: {x_3d.shape}\n")

# ---------------------------------------------------------
# EXAMPLE 1: torch.split on dim=1 (Sequence)
# ---------------------------------------------------------
# We want to split the sequence (size 4) into chunks of size 3.
# This will leave a smaller leftover chunk at the end.
print("--- torch.split (Size-First) ---")
split_pieces = torch.split(x_3d, split_size_or_sections=3, dim=1)

for i, p in enumerate(split_pieces):
    print(f"Split Piece {i} shape: {p.shape}")

print("\n")

# ---------------------------------------------------------
# EXAMPLE 2: torch.chunk on dim=2 (Features)
# ---------------------------------------------------------
# We want exactly 3 equal chunks from the Features dimension (size 6).
# 6 / 3 = 2, so each chunk will have 2 features.
print("--- torch.chunk (Quantity-First) ---")
chunk_pieces = torch.chunk(x_3d, chunks=3, dim=2)

for i, p in enumerate(chunk_pieces):
    print(f"Chunk Piece {i} shape: {p.shape}")
```

```python
matrix = torch.arange(12).reshape(3, 4)
print(f"Original shape: {matrix.shape}\n")

# Unbind along the rows (dim=0)
rows = torch.unbind(matrix, dim=0)
cols = torch.unbind(matrix, dim=1)
rows, cols
```

```python
# Shape: [Batch=2, Sequence=4, Features=6]
x_3d = torch.arange(48).reshape(2, 4, 6)
print(f"Original shape: {x_3d.shape}\n")

# Unbind along the Batch dimension (dim=0)
batch_items = torch.unbind(x_3d, dim=0)

print(f"Number of items in batch: {len(batch_items)}\n")
for i, item in enumerate(batch_items):
    print(f"Batch item {i} shape: {item.shape}")
```
